```python
#Step 1
#Составим систему линейных уравнений

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate, linalg

X_axe = np.arange(1, 16, 1)
Y_axe = np.sin(X_axe/5)*np.exp(X_axe/10)+5*np.exp(-X_axe/2)

A = np.array([[1,1], [1,15]])
b = np.array([float(Y_axe[0]),float(Y_axe[14])])
b = b.reshape((2,1))
print(A)

result = linalg.solve(A, b)
plt.plot(X_axe, Y_axe)
plt.plot([1,15], b)
plt.show()
print(result)
```

```python
#Step 2
#Повторим те же шаги для многочлена второй степени

A = np.array([[1,1,1],[1,8,8**2],[1,15,15**2]])
b = np.array([float(Y_axe[0]),float(Y_axe[9]), float(Y_axe[14])])
b = b.reshape(3,1)
print(A)

result = linalg.solve(A, b)
plt.plot(X_axe, Y_axe)
plt.plot([1,8,15], b)
plt.show()
print(result)
```

```python
#Step 3
#Повторим те же шаги для многочлена третьей степени

A = np.array([[1,1,1,1],[1,4,4**2,4**3],[1,10,10**2,10**3],[1,15,15**2,15**3]])
b = np.array([float(Y_axe[0]),float(Y_axe[3]),float(Y_axe[9]), float(Y_axe[14])])
b = b.reshape(4,1)
print(A)

result = linalg.solve(A, b)
plt.plot(X_axe, Y_axe)
plt.plot([1,4,10,15], b)
plt.show()
print("The results: \n", "w_0: ", result[0], "\n",  "w_1: ", result[1], "\n", "w_2: ", result[2], "\n", "w_3: ", result[3])
```
